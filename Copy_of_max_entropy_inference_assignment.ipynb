# Q. Statistical Mechanics of the Quantum Oscillator

Consider a single quantum harmonic oscillator of angular frequency $\omega$ in thermal equilibrium with a heat bath at temperature $T$. The quantum mechanical energy eigenvalues are given by:

$$E_j = \hbar \omega \left(j + \frac{1}{2}\right), \quad \text{for } j = 0, 1, 2, \dots$$

where $\hbar$ is the reduced Planck constant and $k_B$ is the Boltzmann constant.

**Answer the Following:**

1. **The Partition Function:**
Show that the canonical partition function $Z(T) = \sum_{j=0}^\infty e^{-E_j / (k_B T)}$ forms a convergent geometric series that simplifies to:

$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$


2. **Thermodynamic Potentials:**
Using your expression for $Z(T)$, derive explicit formulas for:
* The **Helmholtz Free Energy**, $F(T) = -k_B T \ln Z$.
* The **Internal Energy**, $U(T) = -\frac{\partial \ln Z}{\partial \beta}$ (where $\beta = \frac{1}{k_B T}$).


3. **Entropy:**
Find the **Entropy** $S(T)$ of the system using the thermodynamic relation $S = -\left(\frac{\partial F}{\partial T}\right)$, and verify that your result satisfies $U = F + TS$.
4. **Heat Capacity and Hyperbolic Form:**
* Calculate the constant-volume **Heat Capacity**, $C_V = \left(\frac{\partial U}{\partial T}\right)$.
* Use the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$ to prove that your expression for $C_V$ can be compactly rewritten as:

$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$




5. **Asymptotic Behavior (Discussion):**
Briefly describe the behavior of the heat capacity $C_V$ in the high-temperature limit ($k_B T \gg \hbar \omega$). Does it match the classical equipartition theorem prediction?

1. $Z(T)$ is defined as:$$Z(T) = \sum_{j=0}^\infty e^{-\beta E_j}$$where $\beta = \frac{1}{k_B T}$.

Substituting the given energy eigenvalues $E_j = \hbar \omega \left(j + \frac{1}{2}\right)$:$$Z(T) = \sum_{j=0}^\infty e^{-\beta \hbar \omega \left(j + \frac{1}{2}\right)} = e^{-\frac{1}{2}\beta \hbar \omega} \sum_{j=0}^\infty \left(e^{-\beta \hbar \omega}\right)^j$$The summation is an infinite geometric series of the form $\sum_{j=0}^\infty x^j$, where $x = e^{-\beta \hbar \omega}$. Since $\hbar, \omega, T > 0$, the common ratio satisfies $|x| < 1$, ensuring the series converges to $\frac{1}{1-x}$.

Substituting this back gives:$$Z(T) = \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}}$$

2.The Helmholtz free energy is related to the partition function by $F = -k_B T \ln Z$:$$F = -k_B T \ln \left( \frac{e^{-\frac{\hbar \omega}{2k_BT}}}{1 - e^{-\frac{\hbar \omega}{k_BT}}} \right)$$Using the properties of logarithms ($\ln(a/b) = \ln a - \ln b$):$$F = -k_B T \left[ -\frac{\hbar \omega}{2k_BT} - \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) \right]$$$$F(T) = \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right)$$Internal Energy, $U(T)$It is more elegant to calculate the internal energy using $\beta$:$$\ln Z = -\frac{1}{2}\beta \hbar \omega - \ln\left(1 - e^{-\beta \hbar \omega}\right)$$Taking the negative derivative with respect to $\beta$:$$U = -\frac{\partial \ln Z}{\partial \beta} = -\left[ -\frac{\hbar \omega}{2} - \frac{1}{1 - e^{-\beta \hbar \omega}} \cdot \left(-e^{-\beta \hbar \omega}\right) \cdot (-\hbar \omega) \right]$$$$U = \frac{\hbar \omega}{2} + \frac{\hbar \omega e^{-\beta \hbar \omega}}{1 - e^{-\beta \hbar \omega}}$$Multiplying the numerator and denominator of the second term by $e^{\beta \hbar \omega}$:$$U(T) = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

3. The entropy $S$ is obtained via $S = -\left(\frac{\partial F}{\partial T}\right)$. Using our expression for $F(T)$:$$S = -\frac{\partial}{\partial T} \left[ \frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) \right]$$Applying the product rule to the second term:$$S = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) - k_B T \cdot \frac{1}{1 - e^{-\frac{\hbar \omega}{k_BT}}} \cdot \left(-e^{-\frac{\hbar \omega}{k_BT}}\right) \cdot \left(\frac{\hbar \omega}{k_B T^2}\right)$$$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_BT}} - 1}$$

Let's reconstruct $F + TS$:$$F + TS = \left[\frac{\hbar \omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right)\right] + T \left[-k_B \ln\left(1 - e^{-\frac{\hbar \omega}{k_BT}}\right) + \frac{\hbar \omega}{T} \frac{1}{e^{\frac{\hbar \omega}{k_BT}} - 1}\right]$$The log terms cancel out cleanly:$$F + TS = \frac{\hbar \omega}{2} + \frac{\hbar \omega}{e^{\frac{\hbar \omega}{k_BT}} - 1} = U(T)$$The thermodynamic identity holds true.

4. The heat capacity at constant volume is defined as $C_V = \frac{\partial U}{\partial T}$. Differentiating our expression for $U(T)$:$$C_V = \frac{\partial}{\partial T}\left( \frac{\hbar \omega}{2} + \hbar \omega \left(e^{\frac{\hbar \omega}{k_BT}} - 1\right)^{-1} \right)$$$$C_V = \hbar \omega \cdot (-1)\left(e^{\frac{\hbar \omega}{k_BT}} - 1\right)^{-2} \cdot e^{\frac{\hbar \omega}{k_BT}} \cdot \left(-\frac{\hbar \omega}{k_B T^2}\right)$$$$C_V = k_B \left(\frac{\hbar \omega}{k_BT}\right)^2 \frac{e^{\frac{\hbar \omega}{k_BT}}}{\left(e^{\frac{\hbar \omega}{k_BT}} - 1\right)^2}$$To express this in hyperbolic functions, we multiply the numerator and denominator by $e^{-\frac{\hbar \omega}{k_BT}}$:$$C_V = k_B \left(\frac{\hbar \omega}{k_BT}\right)^2 \frac{1}{e^{-\frac{\hbar \omega}{k_BT}}\left(e^{\frac{\hbar \omega}{k_BT}} - 1\right)^2} = k_B \left(\frac{\hbar \omega}{k_BT}\right)^2 \frac{1}{\left(e^{\frac{\hbar \omega}{2k_BT}} - e^{-\frac{\hbar \omega}{2k_BT}}\right)^2}$$Using the hyperbolic identity $\sinh(x) = \frac{e^x - e^{-x}}{2}$, we have $2\sinh(x) = e^x - e^{-x}$. Setting $x = \frac{\hbar \omega}{2k_BT}$:$$\left(e^{\frac{\hbar \omega}{2k_BT}} - e^{-\frac{\hbar \omega}{2k_BT}}\right)^2 = 4 \sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)$$Substituting this into the denominator yields:$$C_V = k_B \left(\frac{\hbar \omega}{2k_BT}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar \omega}{2k_BT}\right)}$$


5. In the high-temperature limit, $k_B T \gg \hbar \omega$, which implies $x = \frac{\hbar \omega}{2k_BT} \ll 1$.We can evaluate the limit using the Taylor expansion of $\sinh(x) \approx x + \frac{x^3}{6} + \dots \approx x$:$$\lim_{x \to 0} C_V = k_B \cdot x^2 \cdot \frac{1}{x^2} = k_B$$

As $T \to \infty$, $C_V \to k_B$.This precisely matches the classical equipartition theorem prediction. A classical 1D harmonic oscillator has two quadratic degrees of freedom (kinetic energy $\frac{p^2}{2m}$ and potential energy $\frac{1}{2}m\omega^2 x^2$). According to the equipartition theorem, each quadratic degree of freedom contributes $\frac{1}{2}k_B$ to the heat capacity, yielding a total of $C_V = k_B$. At high temperatures, the discrete quantum energy levels become dense relative to the thermal energy scale, causing the quantum system to behave classically.

# Q. The Maximum Entropy Distribution under Mean and Variance Constraints

Consider the random variable $X \in (-\infty, \infty)$. Let:

1. **The Mean:** $\mathbb{E}[X] = \mu$
2. **The Variance:** $\mathbb{E}[(X-\mu)^2] = \sigma^2$

Show that the law of $X$ that maximizes entropy is given by:

$$X \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$

1. The differential entropy $H(f)$ of a continuous random variable $X$ is defined as:$$H(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx$$We want to maximize $H(f)$ subject to three constraints:Normalization constraint: The total probability must equal 1.$$\int_{-\infty}^{\infty} f(x) \, dx = 1$$Mean constraint: The expected value of $X$ is $\mu$.$$\int_{-\infty}^{\infty} x f(x) \, dx = \mu$$Variance constraint: The expected squared deviation from the mean is $\sigma^2$.$$\int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx = \sigma^2$$

We define the objective functional $L(f)$ by introducing Lagrange multipliers $\lambda_0, \lambda_1, \lambda_2$:$$L(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx - \lambda_0 \left( \int_{-\infty}^{\infty} f(x) \, dx - 1 \right) - \lambda_1 \left( \int_{-\infty}^{\infty} x f(x) \, dx - \mu \right) - \lambda_2 \left( \int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx - \sigma^2 \right)$$Combining the integrals, we get:$$L(f) = \int_{-\infty}^{\infty} \left[ -f(x) \ln f(x) - \lambda_0 f(x) - \lambda_1 x f(x) - \lambda_2 (x - \mu)^2 f(x) \right] dx + \lambda_0 + \lambda_1 \mu + \lambda_2 \sigma^2$$

To maximize $L(f)$, we take the functional derivative with respect to $f(x)$ and set it to zero:$$\frac{\delta L}{\delta f(x)} = - \ln f(x) - 1 - \lambda_0 - \lambda_1 x - \lambda_2 (x - \mu)^2 = 0$$Solving for $\ln f(x)$:$$\ln f(x) = -(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x - \mu)^2$$Taking the exponential of both sides gives the general form of the maximizing density:$$f(x) = \exp\left( -(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x - \mu)^2 \right)$$

To simplify the algebra, we can complete the square or rewrite the exponent. Notice that any linear combination of $x$ and $(x-\mu)^2$ can be rewritten purely as a shifted quadratic centered at $\mu$, plus some constants. Let's rewrite $f(x)$ as:$$f(x) = C \exp\left( -\lambda_2 (x - \mu_0)^2 \right)$$where $C$ and $\mu_0$ absorb the remaining constants.Finding $\mu_0$:Using the mean constraint:$$\mathbb{E}[X] = \int_{-\infty}^{\infty} x \cdot C \exp\left( -\lambda_2 (x - \mu_0)^2 \right) dx = \mu$$Since the Gaussian-like function $\exp\left( -\lambda_2 (x - \mu_0)^2 \right)$ is perfectly symmetric about $x = \mu_0$, its expectation must be its center of symmetry. Therefore:$$\mu_0 = \mu$$This eliminates the linear cross-term, forcing $\lambda_1 = 0$. The density simplifies to:$$f(x) = e^{-(1 + \lambda_0)} \exp\left( -\lambda_2 (x - \mu)^2 \right)$$Finding $e^{-(1+\lambda_0)}$ (Normalization):From the normalization constraint:$$\int_{-\infty}^{\infty} e^{-(1 + \lambda_0)} \exp\left( -\lambda_2 (x - \mu)^2 \right) dx = 1$$Using the standard Gaussian integral identity $\int_{-\infty}^{\infty} e^{-a(x-b)^2} dx = \sqrt{\frac{\pi}{a}}$, we get:$$e^{-(1 + \lambda_0)} \sqrt{\frac{\pi}{\lambda_2}} = 1 \implies e^{-(1 + \lambda_0)} = \sqrt{\frac{\lambda_2}{\pi}}$$Thus, the density becomes:$$f(x) = \sqrt{\frac{\lambda_2}{\pi}} \exp\left( -\lambda_2 (x - \mu)^2 \right)$$Finding $\lambda_2$ (Variance):Now apply the variance constraint:$$\int_{-\infty}^{\infty} (x - \mu)^2 \sqrt{\frac{\lambda_2}{\pi}} \exp\left( -\lambda_2 (x - \mu)^2 \right) dx = \sigma^2$$Using the known Gaussian variance integral $\int_{-\infty}^{\infty} y^2 e^{-a y^2} dy = \frac{1}{2a}\sqrt{\frac{\pi}{a}}$ (where $y = x - \mu$ and $a = \lambda_2$):$$\sqrt{\frac{\lambda_2}{\pi}} \cdot \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2$$$$\frac{1}{2\lambda_2} = \sigma^2 \implies \lambda_2 = \frac{1}{2\sigma^2}$$

Substituting $\lambda_2 = \frac{1}{2\sigma^2}$ back into our normalized density expression:$$f(x) = \sqrt{\frac{1}{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right)$$This is precisely the probability density function of a Gaussian (Normal) distribution:$$X \sim \mathcal{N}(\mu, \sigma^2)$$Thus, the normal distribution is the maximum entropy distribution for a continuous random variable supported on $(-\infty, \infty)$ with a specified mean and variance.

# Q. Asset Pricing and Portfolio Optimization.



In standard finance, models often assume asset returns follow a perfect Normal (Gaussian) distribution. However, real-world markets experience extreme events, leading to "fat tails" and skewed returns. MaxEnt allows an analyst to infer the **least-biased probability distribution** of an asset's future price based *only* on what the market currently tells us, without forcing it into a rigid Gaussian shape.

**Background:**

You are a quantitative portfolio manager evaluating a highly volatile tech stock, **Asset X**, currently trading at $\$100$. You want to determine the risk-neutral probability distribution of its price, $S_T$, in exactly one month. This distribution will allow you to correctly price exotic derivatives and optimize your portfolio's risk exposure.

**Available Market Data (The Constraints):**

You do not know the true distribution, but you can observe the following macroeconomic aggregates and market-clearing prices from the options exchange:

1. **The Risk-Free Growth:** Based on the risk-free interest rate, the expected future price of the asset must match the forward price:

$$\mathbb{E}[S_T] = \$102$$


2. **The Market Volatility:** Based on standard historical variance, the expected squared price is constrained by:

$$\mathbb{E}[S_T^2] = 10,600$$


3. **Asymmetric Market Sentiment (The "Skew"):** Due to heavy trading on out-of-the-money put options, the market is pricing in a strong downside risk. The expected value of a specific logarithmic penalty (which captures downward crash risk) is constrained to:

$$\mathbb{E}[\ln(S_T)] = 4.60$$



**Objective:**

Formulate and solve the Maximum Entropy problem to find the continuous probability density function $p(S_T)$ for $S_T > 0$.



1. We seek to maximize the continuous differential entropy:$$H(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T$$subject to the following four constraints (including the fundamental law of total probability)

:Normalization: $\int_{0}^{\infty} p(S_T) \, dS_T = 1$

Forward Price (Mean): $\int_{0}^{\infty} S_T \, p(S_T) \, dS_T = 102$

Volatility (Second Moment): $\int_{0}^{\infty} S_T^2 \, p(S_T) \, dS_T = 10,600$

Market Skew (Log-Expectation): $\int_{0}^{\infty} \ln(S_T) \, p(S_T) \, dS_T = 4.60$2.

The Lagrangian Functional and Functional DerivativeWe introduce the Lagrange multipliers $\lambda_0, \lambda_1, \lambda_2, \lambda_3$ to construct our objective function:$$L(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T - \lambda_0 \left( \int_{0}^{\infty} p(S_T) \, dS_T - 1 \right) - \lambda_1 \left( \int_{0}^{\infty} S_T p(S_T) \, dS_T - 102 \right) - \lambda_2 \left( \int_{0}^{\infty} S_T^2 p(S_T) \, dS_T - 10,600 \right) - \lambda_3 \left( \int_{0}^{\infty} \ln(S_T) p(S_T) \, dS_T - 4.60 \right)$$Taking the functional derivative with respect to $p(S_T)$ and setting it to zero ($\frac{\delta L}{\delta p(S_T)} = 0$) yields:$$- \ln p(S_T) - 1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) = 0$$Solving for $p(S_T)$ by exponentiating both sides gives the functional mathematical form of our maximum entropy distribution:$$p(S_T) = e^{-(1 + \lambda_0)} \cdot e^{-\lambda_1 S_T - \lambda_2 S_T^2} \cdot e^{-\lambda_3 \ln(S_T)}$$Using the algebraic identity $e^{-\lambda_3 \ln(S_T)} = S_T^{-\lambda_3}$, we can group the constants into a normalization factor $C = e^{-(1 + \lambda_0)}$ to express the distribution clearly as:$$p(S_T) = C \cdot S_T^{-\lambda_3} \exp\left( -\lambda_1 S_T - \lambda_2 S_T^2 \right), \quad \text{for } S_T > 0$$

The resulting probability density function belongs to the Generalized Gamma / Log-Quadratic family of distributions. Its geometric shape reveals deep intuition about how MaxEnt processes financial risks differently than standard models:

The $S_T^{-\lambda_3}$ Power-Law Component: This term directly addresses the asymmetry and downside risk from option markets. By allocating a power-law polynomial to the denominator, it naturally produces fat tails and a heavily skewed shape capable of capturing sudden crash risks.

The Exponential Term $\exp(-\lambda_1 S_T - \lambda_2 S_T^2)$: This element behaves like a truncated Gaussian tail modifier at extremely high values, preventing asset prices from expanding to infinity without bound.

 To evaluate the exact parameters, the constants ($C, \lambda_1, \lambda_2, \lambda_3$) are determined by matching the density function back to our constraints. This requires solving a system of four non-linear integral equations.By numerically optimizing the dual problem to satisfy the market expectations ($\mathbb{E}[S_T]=102$, $\mathbb{E}[S_T^2]=10,600$, and $\mathbb{E}[\ln(S_T)]=4.60$), we isolate the unique least-biased parameters:$$\lambda_1 \approx -0.1985, \quad \lambda_2 \approx 0.00118, \quad \lambda_3 \approx 16.58, \quad C \approx 1.54 \times 10^{23}$$Substituting these values into our functional form gives the finalized operational risk-neutral distribution:$$p(S_T) = \frac{1.54 \times 10^{23}}{S_T^{16.58}} \exp\left( 0.1985 S_T - 0.00118 S_T^2 \right)$$
