*   Let $\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n$ be an independent and identically distributed sample of size $n$ from a multivariate population with mean vector $\boldsymbol{\mu}$ and covariance matrix $\boldsymbol{\Sigma}$. The sample mean vector is:
\begin{equation}
\widehat{\boldsymbol{\mu}}_n = \frac{1}{n}\sum_{i=1}^n \mathbf{x}_i
\end{equation}

The Maximum Likelihood Estimator (MLE) for the covariance matrix is given by:
\begin{equation}
\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)^T
\end{equation}

Decomposing the center around the true mean $\boldsymbol{\mu}$:
\begin{equation}
\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n = (\mathbf{x}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})
\end{equation}

Substituting this into the MLE formula and expanding the outer product yields:
\begin{align}
\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} &= \frac{1}{n}\sum_{i=1}^n \Big[ (\mathbf{x}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \Big] \Big[ (\mathbf{x}_i - \boldsymbol{\mu}) - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \Big]^T \\
&= \frac{1}{n}\sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})(\mathbf{x}_i - \boldsymbol{\mu})^T - (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T
\end{align}

Taking the mathematical expectation of both sides:
\begin{equation}
\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{1}{n}\sum_{i=1}^n \mathbb{E}\left[(\mathbf{x}_i - \boldsymbol{\mu})(\mathbf{x}_i - \boldsymbol{\mu})^T\right] - \mathbb{E}\left[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T\right]
\end{equation}

Using the identities $\mathbb{E}\left[(\mathbf{x}_i - \boldsymbol{\mu})(\mathbf{x}_i - \boldsymbol{\mu})^T\right] = \boldsymbol{\Sigma}$ and $\mathbb{E}\left[(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu})^T\right] = \frac{1}{n}\boldsymbol{\Sigma}$:
\begin{align}
\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] &= \frac{1}{n}(n\boldsymbol{\Sigma}) - \frac{1}{n}\boldsymbol{\Sigma} \\
&= \boldsymbol{\Sigma} - \frac{1}{n}\boldsymbol{\Sigma} \\
&= \frac{n-1}{n}\boldsymbol{\Sigma}
\end{align}


* To eliminate the operational bias, we scale the MLE estimator by Bessel's correction factor $\frac{n}{n-1}$:
\begin{equation}
\mathbf{S} = \frac{n}{n-1}\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}
\end{equation}

Expanding this gives the unique empirical sample covariance matrix:
\begin{equation}
\mathbf{S} = \frac{1}{n-1}\sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)(\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n)^T
\end{equation}

Taking expectations confirms that $\mathbf{S}$ is completely unbiased:
\begin{equation}
\mathbb{E}[\mathbf{S}] = \frac{n}{n-1} \left( \frac{n-1}{n}\boldsymbol{\Sigma} \right) = \boldsymbol{\Sigma}
\end{equation}

* Under the framework of continuous structural health monitoring, hypothesis testing rules are declared as:

 $H_1$: The structural asset exhibits an anomalous or damaged state.


Type I Error ($\alpha$): The false alarm rate, occurring when a healthy asset is classified as anomalous.
\begin{equation}
\alpha = P(\text{Reject } H_0 \mid H_0 \text{ is true})
\end{equation}

Type II Error ($\beta$): The missed detection rate, occurring when a damaged asset fails to trigger the alarm.
\begin{equation}
\beta = P(\text{Fail to reject } H_0 \mid H_1 \text{ is true})
\end{equation}


*    \begin{equation}
\text{Power} = 1 - \beta
\end{equation}
When the significance threshold is set to an ultra-conservative level ($\alpha = 0.0001$), the decision boundary moves deep into the tail of the nominal distribution ($H_0$). This drastically increases the probability of committing a Type II error ($\beta \to 1$). Consequently, the statistical power drops precipitously:
\begin{equation}
\lim_{\alpha \to 0} (1 - \beta) = 0
\end{equation}


The multivariate boundary of healthy operation is defined by the Chi-squared threshold region:
\begin{equation}
\{\mathbf{x} \mid (\mathbf{x} - \widehat{\boldsymbol{\mu}}_n)^T \mathbf{S}^{-1} (\mathbf{x} - \widehat{\boldsymbol{\mu}}_n) \leq \chi^2_{p, 1-\alpha}\}
\end{equation}

The geometric volume $V$ of this hyper-ellipsoid is given by:
\begin{equation}
V = \frac{\pi^{p/2}}{\Gamma\left(\frac{p}{2} + 1\right)} \det(\mathbf{S})^{1/2} \left(\chi^2_{p, 1-\alpha}\right)^{p/2}
\end{equation}

Changing $\alpha$ to $0.0001$ raises the target confidence envelope ($1-\alpha$) to $99.99\%$. This results in a massive surge in the critical value $\chi^2_{p, 1-\alpha}$. Because $V \propto \left(\chi^2_{p, 1-\alpha}\right)^{p/2}$, the geometric volume of the "healthy operation" confidence ellipsoid expands substantially. *   Let $(\mathbf{X}_n)_{n=1}^{\infty}$ be a sequence of random vectors in $\mathbb{R}^d$, and let $(\mathbf{Y}_n)_{n=1}^{\infty}$ be a sequence of random matrices in $\mathbb{R}^{d \times d}$ defined on the same probability space.

Suppose that as $n \to \infty$, $\mathbf{X}_n$ converges in distribution to a random vector $\mathbf{X}$:
\begin{equation}
\mathbf{X}_n \xrightarrow{d} \mathbf{X}
\end{equation}

And suppose that $\mathbf{Y}_n$ converges in probability to a deterministic, constant matrix $\mathbf{A}$:
\begin{equation}
\mathbf{Y}_n \xrightarrow{p} \mathbf{A}
\end{equation}

Then Slutsky's Theorem asserts that:
\begin{align}
\mathbf{X}_n + \mathbf{Y}_n &\xrightarrow{d} \mathbf{X} + \mathbf{A} \\
\mathbf{Y}_n \mathbf{X}_n &\xrightarrow{d} \mathbf{A} \mathbf{X} \\
\mathbf{Y}_n^{-1} \mathbf{X}_n &\xrightarrow{d} \mathbf{A}^{-1} \mathbf{X} \quad (\text{provided } \mathbf{A} \text{ is non-singular})
\end{align}


*  We begin by establishing the primary sequence from the Lindeberg-Lévy Multivariate Central Limit Theorem:
\begin{equation}
\mathbf{X}_n = \sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathbf{Z} \sim \mathscr{N}(\mathbf{0}, \boldsymbol{\Sigma})
\end{equation}

By the Weak Law of Large Numbers (WLLN), the empirical sample covariance matrix $\mathbf{S}$ is a consistent estimator of the latent covariance matrix $\boldsymbol{\Sigma}$:
\begin{equation}
\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma}
\end{equation}

Since the function $g(\mathbf{A}) = \mathbf{A}^{-1/2}$ is continuous over the space of symmetric positive-definite matrices, the Continuous Mapping Theorem guarantees that:
\begin{equation}
\mathbf{Y}_n = \mathbf{S}^{-1/2} \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}
\end{equation}

We define our sequences according to Slutsky's Multiplication Theorem, where $\mathbf{X}_n \xrightarrow{d} \mathbf{Z}$ and $\mathbf{Y}_n \xrightarrow{p} \boldsymbol{\Sigma}^{-1/2}$. Constructing their product yields:
\begin{equation}
\mathbf{Y}_n \mathbf{X}_n = \mathbf{S}^{-1/2} \left[ \sqrt{n}(\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \right] \xrightarrow{d} \boldsymbol{\Sigma}^{-1/2} \mathbf{Z}
\end{equation}

Because $\mathbf{Z}$ is a multivariate normal vector with a covariance profile of $\boldsymbol{\Sigma}$, the transformed limiting distribution evaluates directly to a standard Multivariate Normal identity:
\begin{equation}
\boldsymbol{\Sigma}^{-1/2} \mathbf{Z} \sim \mathscr{N}\left(\mathbf{0}, \boldsymbol{\Sigma}^{-1/2} \boldsymbol{\Sigma} \boldsymbol{\Sigma}^{-1/2}\right) = \mathscr{N}(\mathbf{0}, \mathbf{I})
\end{equation}

Therefore, Slutsky's Theorem establishes the operational tracking pivot:
\begin{equation}
\sqrt{n} \mathbf{S}^{-1/2} (\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu}) \xrightarrow{d} \mathscr{N}(\mathbf{0}, \mathbf{I})
\end{equation}

For a large, finite operational sample snapshot $n$, we isolate the sample mean vector $\widehat{\boldsymbol{\mu}}_n$ by scaling and shifting the distributions:
\begin{equation}
\widehat{\boldsymbol{\mu}}_n - \boldsymbol{\mu} \sim \mathscr{N}\left(\mathbf{0}, \frac{1}{n}\mathbf{S}\right) \implies \widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\boldsymbol{\mu}, \frac{1}{n}\mathbf{S}\right)
\end{equation}

Substituting the localized calibrated nominal mean $\widehat{\boldsymbol{\mu}}_n$ for the latent vector $\boldsymbol{\mu}$ operationalizes the monitoring boundary configuration used in the live dashboard architecture:
\begin{equation}
\widehat{\boldsymbol{\mu}}_n \sim \mathscr{N}\left(\widehat{\boldsymbol{\mu}}_n, \frac{1}{n}\mathbf{S}\right)
\end{equation}

````python
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

# Generating base stationary multivariate normal noise
base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])
# ----------------------------------------------------------------

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation from scratch.
    """
    X = df.to_numpy()
    n, m = X.shape

    # 1. Split df into g equal consecutive blocks
    chunk_size = n // g_chunks

    # 2. Compute the global mean vector
    global_mean = np.mean(X, axis=0)

    # 3. Initialize matrices for Within (W) and Between (B) variations
    B = np.zeros((m, m))
    W = np.zeros((m, m))

    for j in range(g_chunks):
        # Extract chunk data slice
        chunk = X[j * chunk_size : (j + 1) * chunk_size, :]
        n_j = chunk.shape[0]

        # Calculate individual chunk mean vector
        chunk_mean = np.mean(chunk, axis=0)

        # Accumulate Between-Chunk Matrix (B)
        mean_diff = (chunk_mean - global_mean).reshape(-1, 1)
        B += n_j * (mean_diff @ mean_diff.T)

        # Accumulate Within-Chunk Matrix (W)
        for row in chunk:
            dev = (row - chunk_mean).reshape(-1, 1)
            W += dev @ dev.T

    # 4. Compute Wilks' Lambda
    det_W = np.linalg.det(W)
    det_Total = np.linalg.det(B + W)
    wilks_lambda = det_W / det_Total

    # 5. Apply Bartlett’s chi-square approximation
    multiplier = -(n - 1 - (m + g_chunks) / 2)
    chi2_calc = multiplier * np.log(wilks_lambda)

    # 6. Determine Degrees of Freedom and compute p-value
    df_degrees = m * (g_chunks - 1)
    p_value = 1.0 - stats.chi2.cdf(chi2_calc, df_degrees)

    return {
        "Wilks' Lambda": wilks_lambda,
        "Bartlett Chi-Square": chi2_calc,
        "Degrees of Freedom": df_degrees,
        "p-value": p_value
    }

# Execute completed function
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
for key, val in results.items():
    print(f"{key}: {val:.6f}" if isinstance(val, float) else f"{key}: {val}")
````

\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \left( \widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k} \right)^T \left( \widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k} \right)
\end{equation}

Applying the transpose matrix property $(\mathbf{A} + \mathbf{B})^T = \mathbf{A}^T + \mathbf{B}^T$ to the first term yields:
\begin{equation}
\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \left( \mathbf{z}_{i,k}^T \widehat{\mathbf{P}}_k^T + \mathbf{z}_{i,m-k}^T \widehat{\mathbf{P}}_{m-k}^T \right) \left( \widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k} \right)
\end{equation}

Expanding this product via matrix distribution results in four distinct terms:
\begin{align}
\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \; &\mathbf{z}_{i,k}^T \widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \mathbf{z}_{i,k}^T \widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k} \nonumber \\
&+ \mathbf{z}_{i,m-k}^T \widehat{\mathbf{P}}_{m-k}^T \widehat{\mathbf{P}}_k \mathbf{z}_{i,k} + \mathbf{z}_{i,m-k}^T \widehat{\mathbf{P}}_{m-k}^T \widehat{\mathbf{P}}_{m-k} \mathbf{z}_{i,m-k}
\end{align}

We evaluate this expression by leveraging the fundamental algebraic properties of the principal and residual subspaces:

 The individual column bases are orthonormal: $\widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_k = \mathbf{I}_k$ and $\widehat{\mathbf{P}}_{m-k}^T \widehat{\mathbf{P}}_{m-k} = \mathbf{I}_{m-k}$.
 The two subspaces are mutually orthogonal to one another by design: $\widehat{\mathbf{P}}_k^T \widehat{\mathbf{P}}_{m-k} = \mathbf{0}$ and $\widehat{\mathbf{P}}_{m-k}^T \widehat{\mathbf{P}}_k = \mathbf{0}$.


Substituting these structural matrix identities forces the cross-product terms to vanish completely:
$$\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \mathbf{z}_{i,k}^T (\mathbf{I}_k) \mathbf{z}_{i,k} + \mathbf{0} + \mathbf{0} + \mathbf{z}_{i,m-k}^T (\mathbf{I}_{m-k}) \mathbf{z}_{i,m-k}$$

$$= \mathbf{z}_{i,k}^T \mathbf{z}_{i,k} + \mathbf{z}_{i,m-k}^T \mathbf{z}_{i,m-k}$$

Recognizing that the inner product of a vector with itself equals its squared Euclidean norm ($\mathbf{v}^T\mathbf{v} = \|\mathbf{v}\|^2$), we arrive at the final Pythagorean energy decomposition:
\begin{equation}
\|\mathbf{x}_i - \widehat{\boldsymbol{\mu}}_n\|^2 = \|\mathbf{z}_{i,k}\|^2 + \|\mathbf{z}_{i,m-k}\|^2
\end{equation}
This completes the proof that the total energy of an observation is the independent sum of its principal subspace energy and residual subspace energy.


*  To continuously monitor structural assets, data points are projected into a dual-subspace monitoring architecture using two complementary statistical metrics:

Hotelling's $T^2$ statistic is formulated as $T^2_i = \mathbf{z}_{i,k}^T \mathbf{\Lambda}_k^{-1} \mathbf{z}_{i,k}$ and serves the primary function of tracking variance shifts within the principal hyperplane.

The residual $Q$ statistic (Squared Prediction Error) is formulated as $Q_i = \|\mathbf{e}_i\|^2 = \mathbf{z}_{i,m-k}^T \mathbf{z}_{i,m-k}$ and serves the primary function of tracking variance shifts orthogonal to the principal hyperplane.

Most Sensitive Metric: Hotelling's $T^2$ Statistic

Engineering Justification: Severe environmental actions like high wind loads impose large physical forces, inflating the displacement and deflection amplitudes of the structural asset. However, these deformations still follow the structure's normal, healthy degrees of freedom and internal load paths. Because the underlying geometric correlation structure among the sensors remains completely intact, the resulting data points do not drift into the residual space. Instead, they shift outward along the primary principal component axes, triggering a major spike in the $T^2$ statistic while the $Q$ statistic remains stable.



Most Sensitive Metric: Residual $Q$ Statistic (Squared Prediction Error)

Engineering Justification: An internal structural fracture modifies the actual physical properties of the asset, permanently changing how forces distribute through its members. This material degradation breaks the historical sensor cross-correlations established during the baseline calibration period. Because the data pattern no longer matches the nominal model, the incoming measurements can no longer be accurately reconstructed by the primary $k$ principal axes. The data vector branches out, projecting directly into the residual space. This causes an immediate, significant spike in the $Q$ statistic, highlighting a structural shift rather than a standard heavy loading condition.

***Scenario:** You are building a real-time diagnostics layer for a structural component monitored by $m = 4$ sensors. You have a baseline archive of $n = 3,000$ nominal snapshots. **Your Task:**
Complete the Python class method below inside Google Colab. The function must:

1. Standardize and center the incoming realizations.
2. Decompose the unbiased sample covariance matrix $\mathbf{S}$ using `np.linalg.eigh` and sort components in descending order ($\widehat{\lambda}_1 \ge \dots \ge \widehat{\lambda}_m$).
3. Loop through every possible subspace cutoff value $k \in \{1, 2, 3\}$, calculating the sample vector array for Hotelling’s $T^2$ ($T^2_i = \sum_{j=1}^k \frac{z_{i,j}^2}{\widehat{\lambda}_j}$) and the $Q$ statistic ($Q_i = \sum_{j=k+1}^m z_{i,j}^2$).
4. Compute the expected value (empirical mean) of $T^2$ and $Q$ across all samples for each $k$.
5. Return these metrics to verify your theoretical derivations in Part A.



````python
import numpy as np
import pandas as pd

class SHMDiagnosticsPipeline:
    def __init__(self, data: pd.DataFrame):
        """
        Initializes the structural health monitoring diagnostics engine.
        Accepts a dataframe of size (n, m).
        """
        self.X_raw = data.to_numpy()
        self.n, self.m = self.X_raw.shape

    def evaluate_subspace_truncation(self) -> dict:
        """
        Standardizes data, computes the spectral decomposition from scratch,
        and loops through each cutoff k to map out the expected values
        of Hotelling's T^2 and the residual Q statistic.
        """
        # 1. Standardize and center the incoming realizations
        # Note: To match the theoretical setup of standard PCA on a covariance
        # structure, we center the data.
        mean_vector = np.mean(self.X_raw, axis=0)
        X_centered = self.X_raw - mean_vector

        # 2. Compute the unbiased sample covariance matrix S
        S = (X_centered.T @ X_centered) / (self.n - 1)

        # Decompose S using eigh (optimized for symmetric matrices)
        eigenvalues, eigenvectors = np.linalg.eigh(S)

        # np.linalg.eigh returns them in ascending order. Sort in descending order.
        idx = np.argsort(eigenvalues)[::-1]
        lambdas = eigenvalues[idx]
        P = eigenvectors[:, idx]

        # Project centered coordinates into the principal space: Z = X_centered @ P
        Z = X_centered @ P

        results = {}

        # 3. Loop through every possible subspace cutoff value k
        # For m = 4 sensors, the proper truncation limits are k in {1, 2, 3}
        for k in range(1, self.m):
            # Isolate principal coordinates and eigenvalues
            Z_principal = Z[:, :k]
            lambdas_principal = lambdas[:k]

            # Isolate residual coordinates
            Z_residual = Z[:, k:]

            # 4. Calculate tracking vectors for all n snapshots
            # Hotelling's T^2: Sum of squared principal scores divided by their eigenvalues
            T2_vector = np.sum((Z_principal ** 2) / lambdas_principal, axis=1)

            # Residual Q statistic (SPE): Sum of squared residual scores
            Q_vector = np.sum(Z_residual ** 2, axis=1)

            # 5. Compute the empirical expected values (means) across all samples
            mean_T2 = np.mean(T2_vector)
            mean_Q = np.mean(Q_vector)

            results[f"k={k}"] = {
                "Expected_T2": mean_T2,
                "Expected_Q": mean_Q
            }

        return results

# --- Verification & Simulation Wrapper ---
if __name__ == "__main__":
    # Generating synthetic baseline dataset mimicking 4 strain gauge sensors
    np.random.seed(101)
    n_snapshots = 3000
    m_sensors = 4

    synthetic_cov = [
        [0.15, 0.04, 0.02, 0.01],
        [0.04, 0.10, 0.05, 0.02],
        [0.02, 0.05, 0.08, 0.03],
        [0.01, 0.02, 0.03, 0.06]
    ]

    simulated_data = np.random.multivariate_normal(
        mean=[1.2, -0.5, 3.0, 0.8],
        cov=synthetic_cov,
        size=n_snapshots
    )

    df_baseline = pd.DataFrame(
        simulated_data,
        columns=[f'Sensor_Ch{i}' for i in range(1, m_sensors + 1)]
    )

    # Instantiate and run optimization pipeline
    pipeline = SHMDiagnosticsPipeline(df_baseline)
    subspace_analysis = pipeline.evaluate_subspace_truncation()

    # Display results
    print("--- Subspace Energy Metric Breakdown ---")
    for cutoff, metrics in subspace_analysis.items():
        print(f"Truncation Boundary {cutoff}:")
        print(f"  E[T^2] = {metrics['Expected_T2']:.6f}")
        print(f"  E[Q]   = {metrics['Expected_Q']:.6f}")
````

Applying the linearity of the expectation operator and factoring out the deterministic parameter matrix $\boldsymbol{\Lambda}$:
\begin{equation}
\mathbf{R} = \boldsymbol{\Lambda} \mathbb{E}[\mathbf{F}_i \mathbf{F}_i^T] \boldsymbol{\Lambda}^T + \boldsymbol{\Lambda} \mathbb{E}[\mathbf{F}_i \boldsymbol{\epsilon}_i^T] + \mathbb{E}[\boldsymbol{\epsilon}_i \mathbf{F}_i^T] \boldsymbol{\Lambda}^T + \mathbb{E}[\boldsymbol{\epsilon}_i \boldsymbol{\epsilon}_i^T]
\end{equation}

\begin{align}
\mathbf{R} &= \boldsymbol{\Lambda} (\mathbf{I}) \boldsymbol{\Lambda}^T + \boldsymbol{\Lambda} (\mathbf{0}) + (\mathbf{0}) \boldsymbol{\Lambda}^T + \boldsymbol{\Psi} \\

\end{align}

Geometrically, an orthogonal rotation simply alters the coordinate axes within the latent factor subspace without distorting or scaling the space itself. It redistributes individual column loadings to optimize model interpretability, but preserves all vector lengths from the origin, leaving individual sensor communalities completely invariant.

## **Part B: Parametric Factor Partition Engine & UI Architecture**

The objective is to implement an end-to-end data processing and diagnostic visualization pipeline that instantiates a parametric **Factor Analysis (FA)** engine with **Varimax rotation**. The engine must partition empirical observation metrics into a shared structural subspace and a distinct diagonal noise floor, mapping the results onto a highly structured, asymmetric grid interface.

### Develop a Mathematical Engine & Pre-processing Guardrails


Implement an isolated algorithm to structure and decompose empirical data structures according to the following guidelines:

1. **Z-Score Mapping Layer:** Accept an arbitrary data frame and isolate numeric observation columns. Standardize the observations into standard normal space $Z \in \mathbb{R}^{n \times m}$ applying an unbiased sample standard deviation ($ddof=1$). Build an explicit numerical guardrail that catches zero-variance channels ($\sigma_j = 0$) and caps them to a minimum boundary threshold of $10^{-15}$ to prevent divide-by-zero runtime exceptions.
2. **Dimensionality Constraint Layer:** Let $m$ represent the count of isolated sensor features, and $k$ represent the target count of latent physical factors. Program two validation rule exceptions:
* Reject processing if the operational snapshot count ($n$) is less than or equal to the channel count ($m$).
* Throw an descriptive error if the requested latent space dimension ($k$) is not strictly less than the available feature space dimension ($m$).


3. **Information Decomposition Layer:** Fit a Factor Analysis framework utilizing a Maximum Likelihood or Singular Value Decomposition (SVD) algorithm combined with an orthogonal Varimax rotation matrix ($T$). From the fitted parameters, compute the vector arrays for:
* **Communality Vector ($h^2 \in \mathbb{R}^m$):** The total shared variance captured across each channel.
* **Uniqueness Vector ($\varphi^2 \in \mathbb{R}^m$):** The remaining isolated sensor noise floor.

````python
import numpy as np
import pandas as pd
from sklearn.decomposition import FactorAnalysis

def decompose_empirical_data(df: pd.DataFrame, k_factors: int) -> dict:
    """
    Standardizes input data, enforces dimensionality boundaries, and applies
    Factor Analysis with a Varimax rotation to compute Communality and Uniqueness.
    """
    # -------------------------------------------------------------------------
    # 1. Z-Score Mapping Layer with Zero-Variance Guardrails
    # -------------------------------------------------------------------------
    # Extract only numeric columns
    numeric_df = df.select_dtypes(include=[np.number])
    X = numeric_df.to_numpy(dtype=np.float64)

    n, m = X.shape

    # Calculate empirical mean and unbiased sample standard deviation (ddof=1)
    mean_vector = np.mean(X, axis=0)
    std_vector = np.std(X, axis=0, ddof=1)

    # Numerical guardrail: Identify zero-variance/constant sensor channels
    # Elements matching 0.0 are capped at a safe machine-epsilon floor 1e-15
    std_vector_guarded = np.where(std_vector == 0.0, 1e-15, std_vector)

    # Map raw data to standard normal space Z
    Z = (X - mean_vector) / std_vector_guarded

    # -------------------------------------------------------------------------
    # 2. Dimensionality Constraint Layer
    # -------------------------------------------------------------------------
    # Rule Validation 1: Snapshot count must exceed channel feature count
    if n <= m:
        raise ValueError(
            f"Dimensionality Constraint Violated: Snapshot count n ({n}) "
            f"must be strictly greater than feature channel count m ({m})."
        )

    # Rule Validation 2: Latent space k must be strictly less than feature space m
    if k_factors >= m:
        raise ValueError(
            f"Latent Dimension Constraint Violated: Requested factors k ({k_factors}) "
            f"must be strictly less than feature dimensions m ({m})."
        )

    if k_factors <= 0:
        raise ValueError("Latent factors count k must be a positive integer >= 1.")

    # -------------------------------------------------------------------------
    # 3. Information Decomposition Layer (Factor Analysis + Varimax)
    # -------------------------------------------------------------------------
    # Initializing Scikit-Learn's Factor Analysis layer
    # 'varimax' rotation enforces orthogonal coordinate shifts
    fa = FactorAnalysis(
        n_components=k_factors,
        rotation='varimax',
        svd_method='lapack',
        random_state=42
    )

    # Fit the framework on the standardized standard normal data Z
    fa.fit(Z)

    # Extract the Factor Loading Matrix Lambda (m x k)
    # Lambda_jl describes the correlation between sensor j and latent factor l
    lambda_matrix = fa.components_.T

    # Calculate Communality vector (h^2): Row-wise sum of squared factor loadings
    communality = np.sum(lambda_matrix ** 2, axis=1)

    # Extract Uniqueness vector (psi): Specific variance + measurement noise floor
    # Obtained directly via the fitted model parameters
    uniqueness = fa.noise_variance_

    # Pack metrics into structured series mapping back to original columns
    feature_names = numeric_df.columns.tolist()
    communality_series = pd.Series(communality, index=feature_names)
    uniqueness_series = pd.Series(uniqueness, index=feature_names)

    return {
        "Standardized_Data": Z,
        "Factor_Loadings": pd.DataFrame(lambda_matrix, index=feature_names, columns=[f'Factor_{i+1}' for i in range(k_factors)]),
        "Communality": communality_series,
        "Uniqueness": uniqueness_series
    }

# --- Pipeline Execution and Verification Wrapper ---
if __name__ == "__main__":
    # Generating synthetic operational archive mimicking 5 structural sensors
    np.random.seed(42)
    n_snapshots = 1500

    # Seed data with clear multi-variable shared dependencies
    factor1_true = np.random.normal(0, 1, size=n_snapshots)
    factor2_true = np.random.normal(0, 1, size=n_snapshots)

    s1 = 0.8 * factor1_true + np.random.normal(0, 0.2, n_snapshots)
    s2 = 0.7 * factor1_true + np.random.normal(0, 0.3, n_snapshots)
    s3 = 0.9 * factor2_true + np.random.normal(0, 0.1, n_snapshots)
    s4 = 0.6 * factor2_true + np.random.normal(0, 0.4, n_snapshots)
    s5 = np.ones(n_snapshots) * 4.5  # Dead/Constant Channel to test the numerical guardrail

    df_operational = pd.DataFrame({
        'Strain_Ch1': s1, 'Strain_Ch2': s2,
        'Strain_Ch3': s3, 'Strain_Ch4': s4,
        'Dead_Sensor_Ch5': s5
    })

    try:
        print("Executing Information Decomposition Pipeline...")
        # Processing 5 features down to k=2 common underlying factors
        output_metrics = decompose_empirical_data(df_operational, k_factors=2)

        print("\n=== FACTOR LOADINGS MATRIX ===")
        print(output_metrics["Factor_Loadings"].round(4))

        print("\n=== SYSTEM VARIANCE DECOMPOSITION ===")
        variance_tracking = pd.DataFrame({
            "Communality (h^2)": output_metrics["Communality"],
            "Uniqueness (psi^2)": output_metrics["Uniqueness"],
            "Total Sum (h^2 + psi^2)": output_metrics["Communality"] + output_metrics["Uniqueness"]
        })
        print(variance_tracking.round(4))

    except ValueError as err:
        print(f"Pipeline Execution Halted: {err}")
````

### Latent Score Projection

1. **Mathematical State Reconstruction:** Using Thomson’s Minimum Mean Squared Error (MMSE) regression methodology, transform the standard normal snapshot array ($Z$) directly into hidden factor states $F \in \mathbb{R}^{n \times k}$ using the system relation:

$$F = Z R^{-1} \Lambda_{\text{rotated}}$$


2. **Performance Diagnostic Telemetry:** Calculate the empirical variance of the reconstructed factor scores enforcing an unbiased scale correction ($ddof=1$). Print a clean confirmation breakdown out to the logging console that presents the calculated system-wide metrics:

$$\text{Average System Communality \%} = \frac{1}{m}\sum_{j=1}^m h_j^2 \times 100$$


$$\text{Average System Uniqueness \%} = \frac{1}{m}\sum_{j=1}^m \varphi_j^2 \times 100$$

````python
import numpy as np
import pandas as pd

def compute_thomson_telemetry(Z_matrix: np.ndarray, lambda_df: pd.DataFrame, communality_ser: pd.Series, uniqueness_ser: pd.Series) -> dict:
    """
    Transforms standardized standard normal data into latent factor states
    via Thomson's MMSE regression methodology and logs diagnostic telemetry.
    """
    # Isolate dimensionality parameters
    n, m = Z_matrix.shape
    k_factors = lambda_df.shape[1]

    # 1. Compute the empirical correlation matrix R from the standard normal data Z
    # Since Z is standardized, its covariance matrix is the sample correlation matrix
    R = (Z_matrix.T @ Z_matrix) / (n - 1)

    # Extract the factor loading matrix Lambda
    Lambda = lambda_df.to_numpy()

    # 2. Mathematical State Reconstruction: Compute Thomson's Factor Scores
    # System Relation: F = Z * inv(R) * Lambda
    # Using np.linalg.solve is numerically superior and more stable than direct inversion via inv()
    R_inv_Lambda = np.linalg.solve(R, Lambda)
    F_states = Z_matrix @ R_inv_Lambda

    # 3. Performance Diagnostic Telemetry
    # Calculate the empirical variance of the reconstructed factor scores (ddof=1)
    factor_variances = np.var(F_states, axis=0, ddof=1)

    # Calculate average system metrics from the passed series
    avg_communality_pct = np.mean(communality_ser.to_numpy()) * 100
    avg_uniqueness_pct = np.mean(uniqueness_ser.to_numpy()) * 100

    # Print clean confirmation breakdown out to the logging console
    print("==================================================")
    print("      FACTOR ANALYSIS DIAGNOSTIC TELEMETRY        ")
    print("==================================================")
    print(f"Operational Snapshot Size (n)   : {n}")
    print(f"Isolated Sensor Channels  (m)   : {m}")
    print(f"Retained Latent Factors   (k)   : {k_factors}\n")

    print("--- Empirical Variances of Factor Scores (ddof=1) ---")
    for idx, var in enumerate(factor_variances):
        print(f"  Factor {idx + 1} Reconstructed Variance: {var:.6f}")
    print("\n--- Global System Architecture Parsing ---")
    print(f"  Average System Communality % : {avg_communality_pct:.2f}%")
    print(f"  Average System Uniqueness %  : {avg_uniqueness_pct:.2f}%")
    print(f"  Total Shared Target Sum Check: {avg_communality_pct + avg_uniqueness_pct:.2f}%")
    print("==================================================")

    # Format factor scores into a clean DataFrame
    factor_names = [f'Factor_{i+1}' for i in range(k_factors)]
    F_df = pd.DataFrame(F_states, columns=factor_names)

    return {
        "Factor_Scores": F_df,
        "Factor_Variances": factor_variances,
        "Avg_Communality_Pct": avg_communality_pct,
        "Avg_Uniqueness_Pct": avg_uniqueness_pct
    }

# --- Execution using parameters from previous step ---
if __name__ == "__main__":
    # Re-creating dummy variables to mimic output structures of the previous question
    np.random.seed(42)
    n_samples, m_features, k_dim = 1500, 5, 2

    # 1. Simulate standard normal snapshot matrix Z
    Z_simulated = np.random.normal(0, 1, size=(n_samples, m_features))

    # 2. Simulate Factor Loading Matrix Lambda (m x k)
    dummy_loadings = np.array([
        [0.82, 0.05],
        [0.75, -0.12],
        [0.01, 0.89],
        [-0.08, 0.64],
        [0.00, 0.00] # Guardrailed channel
    ])
    feature_labels = ['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3', 'Strain_Ch4', 'Dead_Sensor_Ch5']
    df_lambda = pd.DataFrame(dummy_loadings, index=feature_labels, columns=[f'Factor_{i+1}' for i in range(k_dim)])

    # 3. Simulate Communality and Uniqueness vectors
    h2_sim = np.sum(dummy_loadings ** 2, axis=1)
    phi2_sim = 1.0 - h2_sim
    # Enforce exact bounds on the dead channel for simulation purposes
    h2_sim[-1] = 0.0
    phi2_sim[-1] = 1.0

    ser_h2 = pd.Series(h2_sim, index=feature_labels)
    ser_phi2 = pd.Series(phi2_sim, index=feature_labels)

    # Run telemetry computation
    telemetry_output = compute_thomson_telemetry(Z_simulated, df_lambda, ser_h2, ser_phi2)
````

### Question 1: The Total Variance Illusion (PCA Subplots 1 & 2 vs. FA Subplots 1 & 2)

* **Context:** In your PCA dashboard, `PC 1` captures roughly $45\%$ of the total system variance, and `PC 2` captures over $30\%$, indicating a highly structured, multi-component system. However, looking at the Factor Analysis dashboard, the **Variance Partitioning Stacked Bar** reveals that `Sensor_4` has a **Uniqueness value ($\varphi^2$) exceeding $98\%$**.
* **Core Problem:** 1. What does a Uniqueness value close to $100\%$ tell you about the physical nature of the data coming from `Sensor_4`?
2. Because PCA maximizes total global variance without distinguishing between types of variance, how does the massive localized noise of `Sensor_4` "trick" the PCA engine into inflating its top eigenvalues?
3. Why is this a major risk if an engineer relies *only* on PCA to build an automated plant anomaly detection framework?





1. A Uniqueness value ($\varphi_4^2$) exceeding $98\%$ tells you that Sensor 4 shares effectively zero statistical information with the rest of the asset. In Factor Analysis, a variable's total variance is partitioned into shared structural information (communality, $h^2$) and isolated noise/idiosyncrasies (uniqueness, $\varphi^2$). Physically, a value this high signifies that Sensor 4 is completely decoupled from the underlying mechanical or thermal modes ($f_1$ and $f_2$) driving the plant.Diagnostic Root Causes:Severe Instrumentation Fault: The sensor has suffered a structural failure—such as a broken internal element, a completely severed signal ground, or severe electromagnetic interference (EMI) leaking into its local loop. It is generating massive, localized white noise.Extreme Isolation: If the sensor is physically healthy, it has been placed in a location completely isolated from the plant's core dynamics (e.g., a structural strain gauge mistakenly left hanging unmounted in mid-air). It reads only its own ambient environment.


2.  PCA is an information-blind optimization algorithm. It maximizes total variance ($Var(X_{\text{total}})$), which mathematically includes both structural cross-correlation and isolated random noise. It treats them exactly the same.In your simulated asset, Sensor 4 has an exceptionally large noise contamination scale ($\sigma^2 \approx 2.0$), giving it a variance vastly larger than the healthy channels (Sensors 1, 2, and 3). Here is how that variance profile skews the mathematics:Objective Function Manipulation: The first principal component optimization seeks a projection vector $\mathbf{w}_1$ that maximizes:$$\max_{\mathbf{w}} \mathbf{w}^T \mathbf{S} \mathbf{w}$$Where $\mathbf{S}$ is the sample covariance matrix. Because the diagonal entry corresponding to Sensor 4 is massive ($\approx 2.0$), the mathematical engine can easily maximize this quadratic form simply by aligning $\mathbf{w}_1$ directly with the Sensor 4 axis.Eigenvalue Inflation: The resulting eigenvalue ($\lambda_1$)—which represents the variance captured by PC 1—is heavily padded by this localized noise floor. PCA "looks" highly successful because PC 1 and PC 2 capture $45\%$ and $30\%$ of the variance, but a huge portion of that "captured energy" is actually just the chaotic electrical static of Sensor 4, not a dominant plant process mode.
