```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
X0 = 2
omega = 2

P = 2*np.pi/omega
T = 3*P

# Required step sizes
steps = [40, 160, 2000]

for n in steps:

    dt = P/n

    # Time array
    t = np.arange(0, T + dt, dt)

    # Numerical solution arrays
    u = np.zeros(len(t))
    v = np.zeros(len(t))

    # Initial conditions
    u[0] = X0
    v[0] = 0

    # Forward Euler Method
    for k in range(len(t)-1):
        u[k+1] = u[k] + dt*v[k]
        v[k+1] = v[k] - dt*(omega**2)*u[k]

    # Exact solution
    u_exact = X0*np.cos(omega*t)

    # Plot
    plt.figure(figsize=(10,5))

    plt.plot(t, u, label="Forward Euler")
    plt.plot(t, u_exact, '--', label="Exact Solution")

    plt.title(f"Forward Euler Method (Δt = P/{n})")

    plt.xlabel("Time")

    plt.ylabel("Displacement")

    plt.grid(True)

    plt.legend()

    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
X0 = 2
omega = 2

# Period
P = 2*np.pi/omega

# Long simulation time
T = 50*P

# Step size
dt = P/2000

# Time array
t = np.arange(0, T + dt, dt)

# Initialize arrays
u = np.zeros(len(t))
v = np.zeros(len(t))

# Initial conditions
u[0] = X0
v[0] = 0

# Forward Euler Method
for k in range(len(t)-1):
    u[k+1] = u[k] + dt*v[k]
    v[k+1] = v[k] - dt*(omega**2)*u[k]

# Exact solution
u_exact = X0*np.cos(omega*t)

# Plot
plt.figure(figsize=(12,6))

plt.plot(t, u, label='Forward Euler')
plt.plot(t, u_exact, '--', label='Exact Solution')

plt.title('Forward Euler Method (Δt = P/2000, T = 50P)')

plt.xlabel('Time')

plt.ylabel('Displacement')

plt.grid(True)

plt.legend()

plt.show()
```

The Forward Euler method remains reasonably accurate because a very small step size $$(Δt=P/2000)$$ is used. However, over a long simulation time $$(T=50P)$$, small numerical errors accumulate. As a result, the amplitude of the numerical solution slowly increases compared to the exact solution. This demonstrates that the Forward Euler method is not energy-conserving and becomes less reliable for long-term simulations of oscillatory systems.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
X0 = 2
omega = 2

# Period
P = 2*np.pi/omega

# Simulation time
T = 3*P

# Required step sizes
steps = [40, 160, 2000]

for n in steps:

    # Step size
    dt = P/n

    # Time array
    t = np.arange(0, T + dt, dt)

    # Initialize arrays
    u = np.zeros(len(t))
    v = np.zeros(len(t))

    # Initial conditions
    u[0] = X0
    v[0] = 0

    # Modified Euler Method
    for k in range(len(t)-1):

        # Update velocity first
        v[k+1] = v[k] - dt*(omega**2)*u[k]

        # Use the new velocity to update displacement
        u[k+1] = u[k] + dt*v[k+1]

    # Exact solution
    u_exact = X0*np.cos(omega*t)

    # Plot
    plt.figure(figsize=(10,5))

    plt.plot(t, u, label='Modified Euler')
    plt.plot(t, u_exact, '--', label='Exact Solution')

    plt.title(f'Modified Euler Method (Δt = P/{n})')

    plt.xlabel('Time')
    plt.ylabel('Displacement')

    plt.grid(True)
    plt.legend()

    plt.show()
```

As the time step decreases from $$P/40$$ to $$P/2000$$, the numerical solution obtained using the Modified Euler method becomes increasingly accurate. Unlike the Forward Euler method, the amplitude remains nearly constant throughout the simulation. The method provides a much better approximation of the exact solution, particularly for oscillatory systems, because it is more stable and introduces significantly less numerical error.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
X0 = 2
omega = 2

# Period
P = 2*np.pi/omega

# Long simulation time
T = 50 * P

# Step size
dt = P / 2000

# Time array
t = np.arange(0, T + dt, dt)

# Initialize arrays
u = np.zeros(len(t))
v = np.zeros(len(t))

# Initial conditions
u[0] = X0
v[0] = 0

# Modified Euler Method
for k in range(len(t) - 1):

    # Update velocity
    v[k+1] = v[k] - dt * (omega**2) * u[k]

    # Update displacement using new velocity
    u[k+1] = u[k] + dt * v[k+1]

# Exact solution
u_exact = X0 * np.cos(omega * t)

# Plot
plt.figure(figsize=(12,6))

plt.plot(t, u, label='Modified Euler')
plt.plot(t, u_exact, '--', label='Exact Solution')

plt.title('Modified Euler Method (Δt = P/2000, T = 50P)')
plt.xlabel('Time')
plt.ylabel('Displacement')

plt.grid(True)
plt.legend()

plt.show()
```

Using the Modified Euler method with a very small step size $$(Δt=P/2000)$$ over a long simulation time $$(T=50P)$$, the numerical solution remains very close to the exact solution. The amplitude stays nearly constant throughout the simulation, and only a very small phase error accumulates over time. Unlike the Forward Euler method, the Modified Euler method exhibits good long-term stability and preserves the oscillatory behaviour of the system much more effectively
