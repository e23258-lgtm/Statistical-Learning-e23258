```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal

# --- 1. SETUP DATA ---
# Range for Temperature (X) and Pressure (P)
t = np.linspace(-4, 4, 100)
p = np.linspace(-8, 8, 100)
T_grid, P_grid = np.meshgrid(t, p)

# Define the Joint Density (Bivariate Normal)
# Mean = [0, 0], Covariance = [[1, 0], [0, 4]] since T~N(0,1) and P~N(0,4) are independent
pos = np.dstack((T_grid, P_grid))
rv = multivariate_normal([0, 0], [[1, 0], [0, 4]])
joint_pdf = rv.pdf(pos)

# --- 2. PLOT JOINT DENSITY (Y) ---
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
contour = plt.contourf(T_grid, P_grid, joint_pdf, cmap='viridis', levels=20)
plt.colorbar(contour, label='Density')
plt.title("Joint Density $f_{T,P}(t,p)$ (Full State $Y$)")
plt.xlabel("Temperature Deviation ($T$)")
plt.ylabel("Pressure Deviation ($P$)")

# --- 3. PLOT MARGINAL DENSITY (X) ---
plt.subplot(1, 2, 2)
x_axis = np.linspace(-4, 4, 100)
plt.plot(x_axis, norm.pdf(x_axis, 0, 1), color='darkblue', lw=2)
plt.fill_between(x_axis, norm.pdf(x_axis, 0, 1), alpha=0.3, color='skyblue')
plt.title("Marginal Density $f_X(x)$ (Temperature Only)")
plt.xlabel("Deviation ($x$)")
plt.ylabel("Density")

plt.tight_layout()
plt.show()

# --- 4. CONCEPTUAL SIGMA-ALGEBRA VISUALIZATION ---
plt.figure(figsize=(6, 5))
plt.contour(T_grid, P_grid, joint_pdf, alpha=0.3)
# Draw vertical strips to represent sigma(X)
for val in np.arange(-3, 4, 1):
    plt.axvline(x=val, color='red', linestyle='--', alpha=0.6)
plt.title("Visualizing $\sigma(X)$ as Vertical Strips")
plt.xlabel("Temperature ($T$)")
plt.ylabel("Pressure ($P$)")
plt.legend(['$\sigma(X)$ partitioning'], loc='upper right')
plt.show()
```
